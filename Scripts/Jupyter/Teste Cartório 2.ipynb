```python
from tkinter import *
from tkinter import ttk

lstEstado = estado
lstCidade = []
lstCartorio = []

#Funções Combobox variáveis
def upd_Cidade(*args):
    global lstCidade
    lstCidade = []
    for nomeCartorio in dados:
        if nomeCartorio['uf'] == varEstado.get():
            lstCidade.append(nomeCartorio['city'])
        
    #Ordenar
    lstCidade.sort()
    
    #Remover Valores Repetidos
    lstCidade = list(dict.fromkeys(lstCidade))
    
    #Salvar Consulta
    cbCidade['values'] = lstCidade

def upd_Cartorio(*args):
    global lstCartorio
    lstCartorio = []

    for nomeCartorio in dados:
        if nomeCartorio['city'] == varCidade.get():
            lstCartorio.append(nomeCartorio['name'])
    
    lstCartorio.sort()
    lstCartorio = list(dict.fromkeys(lstCartorio))
    cbCartorio['values'] = lstCartorio

#Funções para ativar busca nas combobox
def check_inputCidade(event):
    value = event.widget.get()

    if value == '':
        cbCidade['values'] = lstCidade
    else:
        data = []
        for item in lstCidade:
            if value.lower() in item.lower():
                data.append(item)

        cbCidade['values'] = data

def check_inputCartorio(event):
    value = event.widget.get()

    if value == '':
        cbCartorio['values'] = lstCartorio
    else:
        data = []
        for item in lstCartorio:
            if value.lower() in item.lower():
                data.append(item)

        cbCartorio['values'] = data

def check_inputEstado(event):
    value = event.widget.get()

    if value == '':
        cbEstado['values'] = lstEstado
    else:
        data = []
        for item in lstEstado:
            if value.lower() in item.lower():
                data.append(item)

        cbEstado['values'] = data

#Função Get Id
def getIdCartorio():
    
    idCartorio = 'Falha ao Gerar ID'
    
    #Salvar Valores
    valorEstado = cbEstado.get().lower()
    valorCidade = cbCidade.get().lower()
    valorCartorio = cbCartorio.get().lower()

    for column in dados:
        if column['uf'].lower() == valorEstado and column['city'].lower() == valorCidade and column['name'].lower() == valorCartorio:
            idCartorio = column['id']

    textBox = Label(root, text=idCartorio)
    textBox.grid(row=5, column=1)

#Iniciar Interface Gráfica
root = Tk()
root.geometry('800x200') 
varEstado = StringVar()
varCidade = StringVar()

#ComboBox Estado
cbEstado = ttk.Combobox(root, values=lstEstado, width=100, textvariable=varEstado)
cbEstado.grid(row=1, column=1, padx=10, pady=10)

#ComboBox Cidade
cbCidade = ttk.Combobox(root, values=lstCidade, width=100, textvariable=varCidade)
cbCidade.grid(row=2, column=1, padx=10, pady=10)

#ComboBox Cartorio
cbCartorio = ttk.Combobox(root, values=lstCartorio, width=100)
cbCartorio.grid(row=3, column=1, padx=10, pady=10)

#Botão Enviar
botao = Button(root, text='Gerar ID', command=lambda:getIdCartorio())
botao.grid(row=4, column=1, padx=10, pady=10)


varEstado.trace('w', upd_Cidade)
varCidade.trace('w', upd_Cartorio)
cbEstado.bind('<KeyRelease>', check_inputEstado)
cbCidade.bind('<KeyRelease>', check_inputCidade)
cbCartorio.bind('<KeyRelease>', check_inputCartorio)

root.mainloop()
```
